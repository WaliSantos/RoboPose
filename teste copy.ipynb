```python
nome_video = 'video1.mp4'  
cap = cv.VideoCapture(nome_video)

if not cap.isOpened():
    print("eroo ao abrir o video")
else:
    total_frames = int(cap.get(cv.CAP_PROP_FRAME_COUNT)) #v todos os frames pra pegar amostras espalhadas
    
    indices_para_analisar = [0, int(total_frames * 0.5), int(total_frames * 0.9)] #pegando inicio meio e fim
    
    #apenas congurao do plot
    fig, axs = plt.subplots(len(indices_para_analisar), 3, figsize=(15, 12))
    plt.subplots_adjust(hspace=0.3) # Espao entre as linhas

    print(f"analisando frames: {indices_para_analisar} de {total_frames} totais")

    for i, idx in enumerate(indices_para_analisar):
        #pula direto para o frame que se quer
        cap.set(cv.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        
        if ret:
            frame_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
            img_float = frame_rgb.astype(np.float32) / 255.0
            soma = img_float.sum(axis=2) + 1e-5
            
            r = img_float[:,:,0] / soma
            g = img_float[:,:,1] / soma
            
            #definindo os limiares
            mask = (r > 0.4) & (g < 0.3)
            mask_vis = mask.astype(np.uint8) * 255

            #configuracao da coluna 1 frame original
            axs[i, 0].imshow(frame_rgb)
            axs[i, 0].set_title(f"Frame {idx} (Original)")
            axs[i, 0].axis('off')
            
            #configuracao da coluna 2 mapa de calor para o vermelho
            im = axs[i, 1].imshow(r, cmap='jet', vmin=0, vmax=1)
            axs[i, 1].set_title("Intensidade Vermelho (r)")
            axs[i, 1].axis('off')
            # Adiciona barra de cor apenas na primeira linha para no poluir
            if i == 0: fig.colorbar(im, ax=axs[i, 1], fraction=0.046, pad=0.04)

            #configuracao da coluna 3 mscara binria
            axs[i, 2].imshow(mask_vis, cmap='gray')
            axs[i, 2].set_title("Mscara")
            axs[i, 2].axis('off')
        else:
            print(f"erro ao ler frame {idx}")

    cap.release()
    plt.show()
```

### Seguindo a etapa 1 do README: Configurao da matriz

```python
#colocando a matriz H que est no pdf. Essa matriz leva de mundo para imagem
H_pdf = np.array([ 
    [0.9236, 0, 1077.2023],
    [0, -0.9102, 1952.5904],
    [0, 0, 1]
])

#precisamos pegar o contrrio da informao da matriz H, ou seja, precisamos fazer imagem -> mundo. (u,v) -> (X,Y)

H_inv = np.linalg.inv(H_pdf) #calculo a inversa da matriz H
print("matriz inversa calculada")
print(H_inv)


def pixel_para_mundo(u, v): #recebo as coordenadas em pixel e devolvo para milimetros
    ponto_img = np.array([u, v, 1]).reshape(3, 1)   #crio o vetor do ponto na imagem [u, v, 1], o reshape  para garantir que o vetor seja uma matriz coluna
                                                    #util e considerabevl ja que a matriz  3x3
    # Aplica a transformao: P_mundo = H_inv * P_img
    ponto_mundo = np.dot(H_inv, ponto_img)#multiplico a matriz inversa pela matriz coluna do ponto na imagem
    
    if ponto_mundo[2] != 0: #normalizao para o ultimo elemento ser 1
        X = ponto_mundo[0]/ponto_mundo[2]
        Y = ponto_mundo[1]/ponto_mundo[2]
        return float(X), float(Y)
    #return 0.0, 0.0

#resumo:
#1 - pego o pixel u,v
#2 - crio a matriz coluna [u,v,1]
#3 - multiplico pela matriz inversa H_inv
#4 - normalizo dividindo por ponto_mundo[2] que  W em [X,Y,W]
#testeando
x_teste, y_teste = pixel_para_mundo(320, 240)
print(f"\nTeste: Pixel (320, 240) corresponde a ({x_teste:.2f}, {y_teste:.2f}) no mundo.")
```

```python
def processar_frame(frame): #passo um frame para ela, ou seja, ela consegue trabalhar imagem por imagem
    """
    Processa um nico frame:
    1. Segmenta o vermelho (Cromaticidade)
    2. Encontra o marcador
    3. Calcula centride e ngulo
    4. Desenha os resultados no frame
    
    Retorna:
        dados: dicionrio com 'x_real', 'y_real', 'u', 'v', 'theta' (ou None se falhar)
        frame_desenhado: imagem com os desenhos (elipse, ponto, texto)
    """
    # 1. Cromaticidade
    img_float = frame.astype(np.float32) / 255.0 #fazendo a converso padro para float
    soma = img_float.sum(axis=2) + 0.00000000000000001 #eu somei um valor muito pequeno s para evitar diviso por zero caso tenha (0,0,0)
    #r = img_float[:,:,1] / soma
    #r = img_float[:,:,2] / soma
    #r = img_float[:,:,3] / soma
    r = img_float[:,:,2] / soma
    g = img_float[:,:,1] / soma
    
    # SEUS LIMIARES CALIBRADOS (Ajuste aqui se precisar)
    mask = (r > 0.4) & (g < 0.3)
    mask_uint8 = (mask * 255).astype(np.uint8)
    
    # 2. Limpeza Morfolgica
    kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
    mask_clean = cv.morphologyEx(mask_uint8, cv.MORPH_OPEN, kernel)
    mask_clean = cv.morphologyEx(mask_clean, cv.MORPH_CLOSE, kernel)
    
    # 3. Encontrar Contornos
    contours, _ = cv.findContours(mask_clean, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    
    dados_robo = None
    frame_out = frame.copy()
    
    if contours:
        # Pega o maior contorno (assume-se que  o rob)
        c = max(contours, key=cv.contourArea)
        
        # Filtra rudos pequenos (rea mnima)
        if cv.contourArea(c) > 50:
            M = cv.moments(c)
            if M['m00'] != 0:
                # Centride na Imagem
                cx = int(M['m10'] / M['m00'])
                cy = int(M['m01'] / M['m00'])
                
                # Orientao (Theta)
                mu20 = M['mu20'] / M['m00']
                mu02 = M['mu02'] / M['m00']
                mu11 = M['mu11'] / M['m00']
                theta_rad = 0.5 * np.arctan2(2 * mu11, mu20 - mu02)
                theta_deg = np.degrees(theta_rad)
                
                # Coordenadas do Mundo (usando a funo da Etapa 1)
                X_real, Y_real = pixel_para_mundo(cx, cy)
                
                # Pacote de dados para retorno
                dados_robo = {
                    'u': cx, 'v': cy,
                    'x_real': X_real, 'y_real': Y_real,
                    'theta': theta_deg
                }
                
                # --- DESENHOS ---
                # Ponto no centro
                cv.circle(frame_out, (cx, cy), 5, (0, 0, 255), -1)
                # Contorno do objeto
                cv.drawContours(frame_out, [c], -1, (0, 255, 0), 2)
                # Texto com a posio real
                texto = f"Pos: ({X_real:.0f}, {Y_real:.0f})mm"
                cv.putText(frame_out, texto, (cx - 50, cy - 20), 
                           cv.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

    return dados_robo, frame_out

print("Funo 'processar_frame' criada com sucesso.")
```

```python
# ==========================================
# 4. LOOP FINAL: PROCESSAMENTO E ANLISE
# ==========================================
import matplotlib.pyplot as plt
from IPython.display import clear_output
import cv2 as cv
import numpy as np

# --- Configurao Inicial ---
# IMPORTANTE: Garanta que H_inv e pixel_para_mundo j foram definidos nas clulas anteriores!
# Substitua pelo nome do seu vdeo
video_path = 'video1.mp4' 
cap = cv.VideoCapture(video_path)

# Listas para guardar os dados da trajetria
traj_X = []      # Posio X no mundo (mm)
traj_Y = []      # Posio Y no mundo (mm)
traj_Theta = []  # ngulo do rob (graus)
traj_u = []      # Pixel u (para desenho)
traj_v = []      # Pixel v (para desenho)

if not cap.isOpened():
    print(f"Erro: No foi possvel abrir '{video_path}'.")
else:
    print(f"Iniciando processamento de '{video_path}'...")

try:
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        frame_count += 1
        
        # --- O "Crebro": Chama nossa funo de processamento ---
        # Certifique-se de ter rodado a clula com 'def processar_frame' antes!
        dados, frame_desenhado = processar_frame(frame)
        
        if dados:
            # Se o rob foi detectado, guardamos os dados
            traj_u.append(dados['u'])
            traj_v.append(dados['v'])
            traj_X.append(dados['x_real'])
            traj_Y.append(dados['y_real'])
            traj_Theta.append(dados['theta'])
            
            # (Opcional) Desenhar o rastro da trajetria no vdeo
            if len(traj_u) > 1:
                # Converte listas de pontos para o formato que o OpenCV gosta
                pts = np.array(list(zip(traj_u, traj_v)), np.int32)
                # Desenha a linha da trajetria (ciano)
                cv.polylines(frame_desenhado, [pts], False, (255, 255, 0), 2)

        # --- Exibio Inteligente ---
        # Mostra apenas a cada 5 ou 10 frames para o notebook no travar/ficar lento
        if frame_count % 10 == 0:
            # Converter BGR (OpenCV) para RGB (Matplotlib)
            frame_rgb = cv.cvtColor(frame_desenhado, cv.COLOR_BGR2RGB)
            
            plt.figure(figsize=(10, 6))
            plt.imshow(frame_rgb)
            plt.title(f"Rastreamento em Tempo Real - Frame {frame_count}")
            plt.axis('off')
            plt.show()
            
            # Limpa a sada anterior para criar animao
            clear_output(wait=True)

except KeyboardInterrupt:
    print("Processamento interrompido pelo usurio.")

finally:
    cap.release()
    print("Vdeo finalizado.")

# ==========================================
# 5. RESULTADOS E GRFICOS (Ps-Processamento)
# ==========================================
if len(traj_X) > 0:
    print(f"Sucesso! {len(traj_X)} pontos rastreados.")
    
    # Grfico 1: Trajetria no Cho (Plano XY)
    plt.figure(figsize=(8, 8))
    plt.plot(traj_X, traj_Y, 'b.-', label='Trajetria Rob', linewidth=1, markersize=3)
    
    # Marcar Incio (Verde) e Fim (Vermelho)
    plt.plot(traj_X[0], traj_Y[0], 'go', label='Incio', markersize=8)
    plt.plot(traj_X[-1], traj_Y[-1], 'rx', label='Fim', markersize=8)
    
    plt.title("Trajetria Recuperada (Mundo Real - mm)")
    plt.xlabel("X (mm)")
    plt.ylabel("Y (mm)")
    plt.legend()
    plt.grid(True)
    plt.axis('equal') # Mantm a proporo 1:1 (fundamental para ver a forma real)
    plt.show()
    
    # Grfico 2: Orientao ao longo do tempo
    plt.figure(figsize=(12, 4))
    plt.plot(traj_Theta, 'r-')
    plt.title("Orientao do Rob (Theta)")
    plt.xlabel("Frame")
    plt.ylabel("ngulo (graus)")
    plt.grid(True)
    plt.show()
    
    # Grfico 3: Posio X e Y separadas
    plt.figure(figsize=(12, 4))
    plt.plot(traj_X, label='Posio X')
    plt.plot(traj_Y, label='Posio Y')
    plt.title("Evoluo das Coordenadas X e Y")
    plt.xlabel("Frame")
    plt.ylabel("Posio (mm)")
    plt.legend()
    plt.grid(True)
    plt.show()

else:
    print("ALERTA: Nenhum objeto foi detectado durante todo o vdeo.")
    print("Verifique os limiares de cor na funo 'processar_frame'.")
```
